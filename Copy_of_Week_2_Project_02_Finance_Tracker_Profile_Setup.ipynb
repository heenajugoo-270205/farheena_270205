## Step 5 – Write the Code

Translate your pseudocode into Python code. Below is one example solution:

Pseudocode:

Start the program

Display theme color options:

Dark colors: Black, Navy, Dark Gray

Light colors: White, Light Blue, Light Gray

Display currency options: USD, EUR, MUR, AUD, GBP

````python
# Finance Tracker Profile Setup Program with Specific Options

# Ask the user for their name
user_name = input("Enter your name: ")

# Provide options for theme color
print("Choose a theme color from the options below:")
print("Dark colors: Black, Navy, Dark Gray")
print("Light colors: White, Light Blue, Light Gray")
theme_color = input("Enter your chosen color: ")

# Provide options for currency
print("Choose your primary currency from the options below:")
print("USD, EUR, MUR, AUD, GBP")
primary_currency = input("Enter your chosen currency: ")

# Provide options for main savings goal
print("Choose your main savings goal from the options below:")
print("New Laptop, Holiday, Car, Emergency Fund, Home Renovation, Education")
main_savings_goal = input("Enter your savings goal: ")

# Display a personalized welcome message
print("\nWelcome to your Personal Finance Tracker,", user_name + "!")
print("Your profile is ready.")
print("The app will use a", theme_color, "theme and all amounts will be in", primary_currency + ".")
print("Let's start working towards your goal of saving for a", main_savings_goal + "!")
````
