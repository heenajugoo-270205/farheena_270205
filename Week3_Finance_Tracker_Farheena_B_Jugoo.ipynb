1. Asks user for an expense amount

````python
print("Expense Classifier")

# Ask user for input
amount = float(input("Enter the expense amount: "))

# Classify the expense
if amount >= 100:
    category = "Major Expense"
    print("This is a major expense. Review your budget carefully.")
elif amount >= 50:
    category = "Moderate Expense"
    print("This is a moderate expense. Keep monitoring your spending.")
elif amount >= 10:
    category = "Minor Expense"
    print("This is a minor expense. Small amounts can add up.")
else:
    category = "Small Purchase"
    print("This is a small purchase with minimal impact.")

# Display result
print("Category:", category)
````

## Part 3: Smart Purchase Advisor (20 minutes)

### The Challenge
Build a program that helps users make smart purchase decisions by considering multiple factors.

### Requirements

Your purchase advisor should:

1. Get purchase amount and user's available money
2. Check if they can afford it (basic affordability)
3. Give different advice based on purchase size
4. Consider their financial situation

### Starter Framework

````python
print("=== Smart Purchase Advisor ===")

# Get financial information
available_money = float(input("How much money do you have available? $"))
purchase_price = float(input("How much does the item cost? $"))
monthly_income = float(input("What's your monthly income? $"))

# Prevent division by zero
if available_money == 0 or monthly_income == 0:
    print("Available money and monthly income must be greater than zero.")
else:
    # Calculate affordability percentage
    affordability_ratio = (purchase_price / available_money) * 100
    income_ratio = (purchase_price / monthly_income) * 100

    # Decision logic
    if purchase_price > available_money:
        advice = "You cannot afford this purchase right now. Consider saving first."
    elif affordability_ratio > 50:
        advice = "This is a large expense relative to your available money. Think carefully before buying."
    elif income_ratio > 30:
        advice = "This purchase is expensive compared to your monthly income. Consider waiting or budgeting."
    elif affordability_ratio > 20:
        advice = "Moderate expense. You can afford it, but spend wisely."
    else:
        advice = "This is a small purchase. Go ahead if you want it!"

    # Output
    print(f"\nAffordability: {affordability_ratio:.1f}% of available money")
    print(f"Relative to monthly income: {income_ratio:.1f}%")
    print("Advice:", advice)
````

### Sample Decision Logic to Implement

- **Can't afford**: "Save up first - you need $X more"
- **Can afford but expensive**: "You can afford it, but it's Y% of your available money. Consider waiting."
- **Reasonable purchase**: "This seems like a reasonable purchase for your budget"
- **Small purchase**: "This is a minor expense - go for it!"

### Bonus Challenges

If you finish early, add these features:

1. **Emergency fund check**: Warn if purchase would drain emergency savings
2. **Category-based advice**: Different advice for necessities vs luxuries
3. **Payment plan calculator**: Show monthly payment options for large purchases

---



````python
print("=== Smart Purchase Advisor ===")

# Get financial information
available_money = float(input("How much money do you have available? $"))
purchase_price = float(input("How much does the item cost? $"))
monthly_income = float(input("What's your monthly income? $"))

# Optional: Emergency fund (bonus)
emergency_savings = float(input("How much is set aside for emergencies? $"))

# Prevent division by zero
if available_money == 0 or monthly_income == 0:
    print("Available money and monthly income must be greater than zero.")
else:
    affordability_ratio = (purchase_price / available_money) * 100
    income_ratio = (purchase_price / monthly_income) * 100

    # Decision logic
    if purchase_price > available_money:
        shortage = purchase_price - available_money
        advice = f"Save up first - you need ${shortage:.2f} more."
    elif affordability_ratio > 50:
        advice = f"You can afford it, but it's {affordability_ratio:.1f}% of your available money. Consider waiting."
    elif affordability_ratio > 20:
        advice = "This seems like a reasonable purchase for your budget."
    else:
        advice = "This is a minor expense - go for it!"

    # Bonus: Emergency fund check
    if (available_money - purchase_price) < emergency_savings:
        advice += "\nWarning: This purchase may drain your emergency savings!"

    # Output
    print(f"\nAffordability: {affordability_ratio:.1f}% of available money")
    print(f"Relative to monthly income: {income_ratio:.1f}%")
    print("Advice:", advice)
````
